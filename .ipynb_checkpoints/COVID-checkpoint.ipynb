```python
import json
import requests
import pandas as pd
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")
%matplotlib inline
```

# States

```python
stateDailyResponse = requests.get("https://covidtracking.com/api/v1/states/daily.json")
print(stateDailyResponse)
stateResponse = requests.get('https://covidtracking.com/api/v1/states/current.json')
print(stateResponse)

states = pd.DataFrame(json.loads(stateResponse.text))
print(len(states))
statesDaily = pd.DataFrame(json.loads(stateDailyResponse.text))
print(len(statesDaily))
states.head()
```

```python
usa = gpd.read_file('tl_2017_us_state/tl_2017_us_state.shp')
usa['state'] = usa['STUSPS']
print(len(usa['state']))
usa.sample(5)
```

```python
ax = usa.plot(figsize=(40,40))
xlim = ([usa.total_bounds[0], -60])
ylim = ([10,  usa.total_bounds[3]])

ax.set_xlim(xlim)
ax.set_ylim(ylim)

plt.show()
```

```python
merged = usa.set_index('state').join(states.set_index('state'))
merged = merged.reset_index()
merged = merged.fillna(0)
merged[['state', 'positive', 'death', 'geometry']].sample(5)
```

```python
fig, ax = plt.subplots(1, figsize=(40, 20))
ax.axis('off')
ax.set_title('Heat Map', fontdict={'fontsize': '40', 'fontweight' : '3'})

ax.set_xlim(xlim)
ax.set_ylim(ylim)

color = 'Oranges'
vmin, vmax = 0, merged['positive'].max()
sm = plt.cm.ScalarMappable(cmap=color, norm=plt.Normalize(vmin=vmin, vmax=vmax))
sm._A = []
cbar = fig.colorbar(sm)
cbar.ax.tick_params(labelsize=20)

merged.plot('positive', cmap=color, linewidth=0.8, ax=ax, edgecolor='0.8', figsize=(40,20))
```

## Colorado

```python
co = statesDaily[statesDaily['state'] == 'CO']
co.loc[:,'date'] = pd.to_datetime(co.loc[:,'date'], format='%Y%m%d').dt.date
co.head()
```

```python
co.sort_values(by='date').plot.bar(x='date', y='positiveIncrease')
```

```python
co.sort_values(by='date').plot(x='date', y='positive')
```
